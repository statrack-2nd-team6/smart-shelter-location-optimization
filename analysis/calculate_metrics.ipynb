```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

df = pd.read_csv("../data/smart_shelter_dataset_final.csv")

X = pd.DataFrame({
    "cai": df["cai"],
    "ridership": np.log1p(df["ridership"]),  
    "dispatch_half": df["dispatch_half"]
})

X = X.dropna()

# 1. 정규화
scaler = MinMaxScaler()
Xn = scaler.fit_transform(X)

# 2. 분산 계산
variances = Xn.var(axis=0)

# 3. 가중치 산출
weights = variances / variances.sum()

weights_df = pd.DataFrame({
    "지표": X.columns,
    "최종 가중치": weights
})

print(weights_df)
cai_weight, passenger_weight, waiting_weight = 0.4, 0.4, 0.2
```

```python
df.head()
cols = ['priority_v1', 'priority_v2', 'priority_v4']

df.drop(cols, axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

cols = ['cai', 'ridership', 'dispatch_half']
scaler = MinMaxScaler()

df[cols] = scaler.fit_transform(df[cols])
```

```python
df['final_metrics'] = df["cai"] * cai_weight + df["ridership"] * passenger_weight + df['dispatch_half'] * waiting_weight
```
